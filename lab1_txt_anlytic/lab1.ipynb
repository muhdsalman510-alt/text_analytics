```python
import PyPDF2

# EXERCISE 1: Extract ALL pages
with open('Business_Proposal.pdf', 'rb') as file:
    reader = PyPDF2.PdfReader(file)
    all_text = ""
    for page in reader.pages:
        text = page.extract_text()
        if text:
            all_text += text + "\n"
    
with open('business_proposal_all.txt', 'w', encoding='utf-8') as output:
    output.write(all_text)

print("✓ Exercise 1 done!")

# EXERCISE 2: Extract ONLY page 2
with open('Business_Proposal.pdf', 'rb') as file:
    reader = PyPDF2.PdfReader(file)
    page_2 = reader.pages[1]  # Page 2 is index 1
    text = page_2.extract_text()
    
with open('business_proposal_page_2.txt', 'w', encoding='utf-8') as output:
    output.write(text)

print("✓ Exercise 2 done!")
```
